```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv('../clean_dssp_csv/1i3n.csv')
df.head(10)
```

```python
data = df[['AA', 'SS']]
data.head(10)
```

```python
seq = data['AA'].values
seq
```

```python
def construct_positional(d, m=5):
    s = m//2
    pad = ['-']*s + list(d) + ['-']*s
    n = len(d)
    rows = []
    
    for i in xrange(s, n+s):
        row = pad[i-s:i+s+1]
        rows.append(row)
    
    return rows
    
```

```python
def construct_positional_header(m=5):
    s = m//2
    return ['AA'+str(i) for i in xrange(-s, s+1)]
```

```python
construct_positional_header()
```

```python
positional = np.asarray(construct_positional(seq))
```

```python
pos_df = pd.DataFrame(positional, columns=construct_positional_header())
pos_df.head(10)
```

```python
full_df = pos_df.join(data['SS'])
```
